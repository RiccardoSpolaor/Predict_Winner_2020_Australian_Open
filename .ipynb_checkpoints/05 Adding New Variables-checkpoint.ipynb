```python
def setTournamentStatistics():
    
    dataframe.insert(dataframe.columns.get_loc('Loser') + 1, column = 'setsTournA', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('setsTournA') + 1, column = 'setsTournB', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('setsTournB') + 1, column = 'setsWRatioTournA', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('setsWRatioTournA') + 1, column = 'setsWRatioTournB', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('setsWRatioTournB') + 1, column = 'gamesTournA', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('gamesTournA') + 1, column = 'gamesTournB', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('gamesTournB') + 1, column = 'gamesWRatioTournA', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('gamesWRatioTournA') + 1, column = 'gamesWRatioTournB', value = np.nan)
    
    setsTournA = []
    setsTournB = []
    setsWRatioTournA = []
    setsWRatioTournB = []
    gamesTournA = []
    gamesTournB = []
    gamesWRatioTournA = []
    gamesWRatioTournB = []
    
    for _, group in dataframe.groupby(['csvID','ATP'], sort = False):
        players_played_sets = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
        players_won_sets = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
        
        players_played_games = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
        players_won_games = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}

        for r in group.itertuples():
            setsTournA += [players_played_sets[r.Winner]]
            gamesTournA += [players_played_games[r.Winner]]
            
            try:
                setsWRatioTournA += [players_won_sets[r.Winner]/players_played_sets[r.Winner]]
            except:
                setsWRatioTournA += [0]
                
            try:
                gamesWRatioTournA += [players_won_games[r.Winner]/players_played_games[r.Winner]]
            except:
                gamesWRatioTournA += [0]
                
            players_played_sets[r.Winner] += (r.Lsets + r.Wsets)
            players_won_sets[r.Winner] += r.Wsets 
            
            players_played_games[r.Winner] += (r.W1 + r.W2 + r.W3 + r.W4 + r.W5 + r.L1 + r.L2 + r.L3 + r.L4 + r.L5)
            players_won_games[r.Winner] += (r.W1 + r.W2 + r.W3 + r.W4 + r.W5)


            setsTournB += [players_played_sets[r.Loser]]
            gamesTournB += [players_played_games[r.Loser]]
            try:
                setsWRatioTournB += [players_won_sets[r.Loser]/players_played_sets[r.Loser]]
            except:
                setsWRatioTournB += [0]
                
            try:
                gamesWRatioTournB += [players_won_games[r.Loser]/players_played_games[r.Loser]]
            except:
                gamesWRatioTournB += [0]
                
            players_played_sets[r.Loser] += (r.Lsets + r.Wsets)
            players_won_sets[r.Loser] += r.Lsets 
            
            players_played_games[r.Loser] += (r.W1 + r.W2 + r.W3 + r.W4 + r.W5 + r.L1 + r.L2 + r.L3 + r.L4 + r.L5)
            players_won_games[r.Loser] += (r.L1 + r.L2 + r.L3 + r.L4 + r.L5)
                                    
    dataframe['setsTournA'] = setsTournA
    dataframe['setsTournB'] = setsTournB
    dataframe['setsWRatioTournA'] = setsWRatioTournA
    dataframe['setsWRatioTournB'] = setsWRatioTournB
    dataframe['gamesTournA'] = gamesTournA
    dataframe['gamesTournB'] = gamesTournB
    dataframe['gamesWRatioTournA'] = gamesWRatioTournA
    dataframe['gamesWRatioTournB'] = gamesWRatioTournA
            
            
setTournamentStatistics()
```

```python
# Itero una volta aggiungendo la feature delle partite giocate da perdente e vincitore finora

def setPlayedMatches ():
    
    dataframe.insert(dataframe.columns.get_loc('Loser') + 1, column = 'PlayedA', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('PlayedA') + 1, column = 'PlayedB', value = np.nan)
    
    playersDictionary = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
    wPlayed = []
    lPlayed = []
    for item in dataframe.itertuples():
        wPlayed += [playersDictionary[item.Winner]]
        lPlayed += [playersDictionary[item.Loser]]
        playersDictionary[item.Loser] += 1
        playersDictionary[item.Winner] += 1
    dataframe['PlayedA'] = wPlayed
    dataframe['PlayedB'] = lPlayed
    
def setPlayedMatchesByCourt ():
    
    dataframe.insert(dataframe.columns.get_loc('MinL') + 1, column = 'PlayedCourtA', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('PlayedCourtA') + 1, column = 'PlayedCourtB', value = np.nan)
    
    dataframe.insert(dataframe.columns.get_loc('PlayedCourtB') + 1, column = 'WonRatioCourtA', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('WonRatioCourtA') + 1, column = 'WonRatioCourtB', value = np.nan)
    
    played_court_A = []
    played_court_B = []
    won_court_A = []
    won_court_B = []
    
    players_played = {p:{'Clay':0,'Hard':0,'Grass':0} for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
    players_won = {p:{'Clay':0,'Hard':0,'Grass':0} for p in dataframe['Winner'].append(dataframe['Loser']).unique()}

    for item in dataframe.itertuples():

        if item.Hard != 0:
            court = 'Hard'
        elif item.Grass != 0:
            court = 'Grass'
        elif item.Clay != 0:
            court = 'Clay'

        played_court_A += [players_played[item.Winner][court]]
        played_court_B += [players_played[item.Loser][court]]

        try:
            won_court_A += [players_won[item.Winner][court]/players_played[item.Winner][court]]
        except:
            won_court_A += [0]
        try:
            won_court_B += [players_won[item.Loser][court]/players_played[item.Loser][court]]
        except:
            won_court_B += [0]

        players_played[item.Winner][court] += 1
        players_played[item.Loser][court] += 1

        players_won[item.Winner][court] += 1

    dataframe['PlayedCourtA'] = played_court_A
    dataframe['PlayedCourtB'] = played_court_B
    dataframe['WonRatioCourtA'] = won_court_A 
    dataframe['WonRatioCourtB'] = won_court_B
```

```python
def getPlayerWinLossRatio ():
    
    dataframe.insert(dataframe.columns.get_loc('PlayedB') + 1, column = 'WinRatioA', value = np.nan)
    dataframe.insert(dataframe.columns.get_loc('WinRatioA') + 1, column = 'WinRatioB', value = np.nan)
    
    wWinRatio = [] 
    lWinRatio = [] 
    
    winnersDictionary = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
    
    for r in dataframe.itertuples():
        if r.PlayedA == 0:
            wPlayed = 1
        else:
            wPlayed = r.PlayedA
            
        if r.PlayedB == 0:
            lPlayed = 1
        else:
            lPlayed = r.PlayedB
            
        
        wWinRatio += [winnersDictionary[r.Winner]/wPlayed]
        lWinRatio += [winnersDictionary[r.Loser]/lPlayed]
        
        winnersDictionary[r.Winner] += 1
        
    dataframe['WinRatioA'] = wWinRatio
    dataframe['WinRatioB'] = lWinRatio
```
