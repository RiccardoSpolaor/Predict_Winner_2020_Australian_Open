```python
# Itero una volta aggiungendo la feature delle partite giocate da perdente e vincitore finora

def setPlayedMatches ():
    playersDictionary = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
    wPlayed = []
    lPlayed = []
    for item in dataframe.itertuples():
        wPlayed += [playersDictionary[item.Winner]]
        lPlayed += [playersDictionary[item.Loser]]
        playersDictionary[item.Loser] += 1
        playersDictionary[item.Winner] += 1
    dataframe['PlayedA'] = wPlayed
    dataframe['PlayedB'] = lPlayed
```

```python
dataframe.insert(dataframe.columns.get_loc('Loser') + 1, column = 'PlayedA', value = np.nan)
dataframe.insert(dataframe.columns.get_loc('PlayedA') + 1, column = 'PlayedB', value = np.nan)

setPlayedMatches()
```

```python
def getPlayerWinLossRatio ():
    
    wWinRatio = [] 
    lWinRatio = [] 
    wLossRatio = [] 
    lLossRatio = [] 
    
    winnersDictionary = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
    losersDictionary = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
    
    for r in dataframe.itertuples():
        if r.PlayedA == 0:
            wPlayed = 1
        else:
            wPlayed = r.PlayedA
            
        if r.PlayedB == 0:
            lPlayed = 1
        else:
            lPlayed = r.PlayedB
            
        
        wWinRatio += [winnersDictionary[r.Winner]/wPlayed]
        lWinRatio += [winnersDictionary[r.Loser]/lPlayed]
        wLossRatio += [losersDictionary[r.Winner]/wPlayed]
        lLossRatio += [losersDictionary[r.Loser]/lPlayed]
        
        winnersDictionary[r.Winner] += 1
        losersDictionary[r.Loser] += 1
        
    dataframe['WinRatioA'] = wWinRatio
    dataframe['WinRatioB'] = lWinRatio
    dataframe['LossRatioA'] = wLossRatio
    dataframe['LossRatioB'] = lLossRatio
```

```python
dataframe.insert(dataframe.columns.get_loc('PlayedB') + 1, column = 'WinRatioA', value = np.nan)
dataframe.insert(dataframe.columns.get_loc('WinRatioA') + 1, column = 'WinRatioB', value = np.nan)
dataframe.insert(dataframe.columns.get_loc('WinRatioB') + 1, column = 'LossRatioA', value = np.nan)
dataframe.insert(dataframe.columns.get_loc('LossRatioA') + 1, column = 'LossRatioB', value = np.nan)
```

```python
def setLastMatchesMean ():
    w1mean_5 = []
    l1mean_5 = []
    playersDictionary = {p:0 for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
    for i,r in dataframe.iterrows():
        dataframe.iloc[i,w1mean_5] = playersDictionary[r['Winner']]
        dataframe.iloc[i,l1mean_5] = playersDictionary[r['Loser']]
        playersDictionary[r['Loser']] = r['L1']
        playersDictionary[r['Winner']] = r['W1']
```

```python
def getPlayerLastStats(p, w):
    if len(p['1'])==0:
        w['5_1Mean'] += [0]
        w['5_2Mean'] += [0]
        w['5_3Mean'] += [0]
        w['5_4Mean'] += [0]
        w['5_5Mean'] += [0]
        w['5_setsMean'] += [0]
    else:
        w['5_1Mean'] += [np.mean(p['1'])]
        w['5_2Mean'] += [np.mean(p['2'])]
        w['5_3Mean'] += [np.mean(p['3'])]
        w['5_4Mean'] += [np.mean(p['4'])]
        w['5_5Mean'] += [np.mean(p['5'])]
        w['5_setsMean'] += [np.mean(p['sets'])]


def setPlayersStatistics():
    
    w5 = {'5_1Mean' : [],
          '5_2Mean' : [],
          '5_3Mean' : [],
          '5_4Mean' : [],
          '5_5Mean' : [],
          '5_setsMean' : []}
    
    l5 = {'5_1Mean' : [],
          '5_2Mean' : [],
          '5_3Mean' : [],
          '5_4Mean' : [],
          '5_5Mean' : [],
          '5_setsMean' : []}
    
    playersDictionary = {p: {'1': [], '2': [], '3': [], '4': [], '5': [], 'sets': []} for p in dataframe['Winner'].append(dataframe['Loser']).unique()}
    
    for r in dataframe.itertuples():
        
        getPlayerLastStats(playersDictionary[r.Winner], w5)
        
        getPlayerLastStats(playersDictionary[r.Loser], l5)
        
        playersDictionary[r.Winner]['1'] =  (playersDictionary[r.Winner]['1'] + [r.W1])[-5:]
        playersDictionary[r.Winner]['2'] = (playersDictionary[r.Winner]['2'] + [r.W2])[-5:]
        playersDictionary[r.Winner]['3'] = (playersDictionary[r.Winner]['3'] + [r.W3])[-5:]
        playersDictionary[r.Winner]['4'] = (playersDictionary[r.Winner]['4'] + [r.W4])[-5:]
        playersDictionary[r.Winner]['5'] = (playersDictionary[r.Winner]['5'] + [r.W5])[-5:]
        playersDictionary[r.Winner]['sets'] = (playersDictionary[r.Winner]['sets'] + [r.Wsets])[-5:]
        
        playersDictionary[r.Loser]['1'] = (playersDictionary[r.Loser]['1'] + [r.L1])[-5:]
        playersDictionary[r.Loser]['2'] = (playersDictionary[r.Loser]['2'] + [r.L2])[-5:]
        playersDictionary[r.Loser]['3'] = (playersDictionary[r.Loser]['3'] + [r.L3])[-5:]
        playersDictionary[r.Loser]['4'] = (playersDictionary[r.Loser]['4'] + [r.L4])[-5:]
        playersDictionary[r.Loser]['5'] = (playersDictionary[r.Loser]['5'] + [r.L5])[-5:]
        playersDictionary[r.Loser]['sets'] = (playersDictionary[r.Loser]['sets'] + [r.Lsets])[-5:]
                
    dataframe['5_1MeanA'] = w5['5_1Mean']
    dataframe['5_2MeanA'] = w5['5_2Mean']
    dataframe['5_3MeanA'] = w5['5_3Mean']
    dataframe['5_4MeanA'] = w5['5_4Mean']
    dataframe['5_5MeanA'] = w5['5_5Mean']
    dataframe['5_setsMeanA'] = w5['5_setsMean']
    dataframe['5_1MeanB'] = l5['5_1Mean']
    dataframe['5_2MeanB'] = l5['5_2Mean']
    dataframe['5_3MeanB'] = l5['5_3Mean']
    dataframe['5_4MeanB'] = l5['5_4Mean']
    dataframe['5_5MeanB'] = l5['5_5Mean']
    dataframe['5_setsMeanB'] = l5['5_setsMean']
```

```python
%%time

dataframe = dataframe.append(pd.DataFrame(
    columns = ['5_1MeanA', '5_2MeanA', '5_3MeanA', '5_4MeanA',
               '5_5MeanA','5_setsMeanA','5_1MeanB', '5_2MeanB', '5_3MeanB', '5_4MeanB',
               '5_5MeanB','5_setsMeanB',]), sort=False)

setPlayersStatistics()
    
```
