## Feature di Confronto
Vengono aggiunte feature di confronto di maggioranza tra le caratteristiche dei due giocatori (es: ***Rank***; ***Pts***).

Le feature di confronto verranno aggiunte anche per le nuove variabili aggiunte.

```python
dataframe['RankA>RankB'] = dataframe.apply(lambda x : int(x['WRank'] > x['LRank']), axis = 1)
dataframe['MaxA>MaxB'] = dataframe.apply(lambda x : int(x['MaxW'] > x['MaxL']), axis = 1)
dataframe['AvgA>AvgB'] = dataframe.apply(lambda x : int(x['AvgW'] > x['AvgL']), axis = 1)
dataframe['PtsA>PtsB'] = dataframe.apply(lambda x : int(x['WPts'] > x['LPts']), axis = 1)
```
